## **Theory Questions**

1. **What is unsupervised learning in the context of machine learning?**

Unsupervised learning in machine learning is a type of algorithm that analyzes unlabeled data to find hidden patterns, structures, and relationships without any prior guidance or labeled examples. Unlike supervised learning, which relies on labeled data to train a model, unsupervised learning explores the data to discover inherent characteristics and insights.

2. **How does K-Means clustering algorithm work?**

K-means clustering is an unsupervised machine learning algorithm that groups data points into k clusters based on similarity. The algorithm iteratively assigns data points to the nearest centroid (cluster center) and recalculates the centroids until convergence, aiming to minimize the within-cluster variance.

3. **Explain the concept of a dendrogram in hierarchical clustering.**

A dendrogram is a tree-like diagram that visualizes the hierarchical relationships between data points in hierarchical clustering. It shows how clusters are merged or divided at different levels of similarity, with the height of the connections representing the distance (or dissimilarity) between the merged clusters.

4. **What is the main difference between K-Means and Hierarchical Clustering?**

The primary difference between K-Means and Hierarchical Clustering lies in how they determine the number of clusters and represent the data. K-Means requires the user to specify the number of clusters (k) beforehand, while Hierarchical Clustering builds a tree-like structure (dendrogram) of clusters without needing a predefined k. K-Means iteratively assigns data points to the nearest cluster center, while Hierarchical Clustering either merges or splits clusters based on proximity.

5. **What are the advantages of DBSCAN over K-Means?**

DBSCAN offers several advantages over K-means, particularly in handling datasets with varying densities, non-spherical clusters, and noise. DBSCAN excels at identifying clusters of arbitrary shapes and sizes, doesn't require specifying the number of clusters beforehand, and is robust to outliers.

6. **When would you use Silhouette Score in clustering?**

The Silhouette Score is used to evaluate the performance of clustering algorithms, specifically by assessing how well each data point fits within its assigned cluster and how distinct it is from other clusters. It's useful for determining the optimal number of clusters and validating the results of clustering.

7. **What are the limitations of Hierarchical Clustering?**

Hierarchical clustering, while useful, has several limitations. It can be computationally expensive and memory-intensive, especially with large datasets, due to the need to calculate and store distance matrices. It's also sensitive to noise and outliers, and can produce suboptimal results due to the irreversible nature of merging or splitting clusters. Additionally, determining the optimal number of clusters from the dendrogram can be subjective.

8. **Why is feature scaling important in clustering algorithms like K-Means?**

Feature scaling is crucial for K-Means and other distance-based clustering algorithms because it ensures that all features contribute equally to the distance calculations. Without scaling, features with larger values or wider ranges can disproportionately influence the clustering process, leading to biased results.

9. **How does DBSCAN identify noise points?**

The algorithm works by defining clusters as dense regions separated by regions of lower density. This approach allows DBSCAN to discover clusters of arbitrary shape and identify outliers as noise.

10. **Define inertia in the context of K-Means.**

It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. A good model is one with low inertia AND a low number of clusters ( K ). However, this is a tradeoff because as K increases, inertia decreases.

11. **What is the elbow method in K-Means clustering?**

The elbow method is a heuristic used to find the optimal number of clusters (k) in K-means clustering. It involves plotting the within-cluster sum of squares (WCSS) for a range of k values and identifying the "elbow" point, where adding more clusters doesn't significantly reduce the WCSS. This point indicates a good balance between minimizing within-cluster variance and the number of clusters.

12. **Describe the concept of "density" in DBSCAN.**

In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), density refers to the concentration of data points within a specified neighborhood. DBSCAN identifies clusters as dense regions of data points separated by regions of lower density. The algorithm uses two parameters: epsilon (Îµ), the radius of the neighborhood, and minPoints, the minimum number of data points required within that radius for a point to be considered a "core point".

13. **Can hierarchical clustering be used on categorical data?**

Yes, hierarchical clustering can be used with categorical data, but it requires careful consideration of how distances are calculated between data points. Standard distance metrics like Euclidean distance are not suitable for categorical data, so alternative methods for measuring dissimilarity are needed.

14. **What does a negative Silhouette Score indicate?**

Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

15. **Explain the term "linkage criteria" in hierarchical clustering.**

In hierarchical clustering, linkage criteria determine how the distance between clusters is calculated when merging them during the clustering process. Essentially, it defines what "close" means when comparing two clusters, guiding the algorithm's merging decisions. Different linkage criteria can significantly impact the shape and structure of the resulting dendrogram and the final clusters formed.

16. **Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**

K-Means clustering can struggle with datasets exhibiting varying cluster sizes or densities because it inherently assumes that all clusters are roughly spherical, have similar sizes, and are equally dense.

17. **What are the core parameters in DBSCAN, and how do they influence clustering?**

DBSCAN, a density-based clustering algorithm, relies on two key parameters: epsilon (Îµ) and MinPts. Epsilon defines the radius within which to search for neighboring points, while MinPts specifies the minimum number of points required to form a dense region. These parameters directly influence the algorithm's ability to identify clusters of varying densities and handle noise in the data.

18. **How does K-Means++ improve upon standard K-Means initialization?**

K-Means++ improves upon standard K-Means by using a smarter initialization strategy for cluster centroids. Instead of randomly selecting initial centroids, K-Means++ selects them in a way that ensures they are well-separated and distributed across the data space, leading to faster convergence and potentially better clustering results.

19. **What is agglomerative clustering?**

Agglomerative clustering is a hierarchical clustering algorithm that takes a "bottom-up" approach, starting with each data point as its own cluster and then iteratively merging the closest pairs of clusters until all points are in a single cluster. It is a popular method for grouping similar data points and visualizing the relationships between clusters using a dendrogram.

20. **What makes Silhouette Score a better metric than just inertia for model evaluation?**

The Silhouette Score is generally a better metric than inertia (also known as within-cluster sum of squares) for evaluating clustering models because it considers both the compactness of clusters and their separation from other clusters. Inertia only measures the compactness of clusters, potentially leading to misleading interpretations when comparing different clustering structures.

21. **Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# Step 2: Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.75, marker='X', label='Centers')
plt.title('K-Means Clustering on Synthetic Data with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

29. **Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
silhouette_scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Display as a bar chart
plt.figure(figsize=(8, 6))
plt.bar(k_values, silhouette_scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k (make_blobs Data)")
plt.grid(axis='y')
plt.show()
```

30. **Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Perform hierarchical clustering using average linkage
linked = linkage(X, method='average')

# Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linked,
           orientation='top',
           labels=iris.target,
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.grid(True)
plt.tight_layout()
plt.show()
```

31. **Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

# 1ï¸â£ Generate synthetic data with overlapping clusters
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

# 2ï¸â£ Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# 3ï¸â£ Visualize with decision boundaries
def plot_decision_boundaries(X, labels, model, title):
    h = 0.1  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    cmap_background = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_points = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, cmap=cmap_background, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap=cmap_points, s=30, edgecolor='k')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.grid(True)
    plt.show()

plot_decision_boundaries(X, y_kmeans, kmeans, "K-Means Clustering with Decision Boundaries")
```

32. **Load the Digits dataset and apply DBSCAN after reducing dimensions with t-SNE. Visualize the results.**

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np

# 1ï¸â£ Load the Digits dataset
digits = load_digits()
X = digits.data
y_true = digits.target

# 2ï¸â£ Reduce dimensions with t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

# 3ï¸â£ Apply DBSCAN
dbscan = DBSCAN(eps=5, min_samples=5)
labels = dbscan.fit_predict(X_tsne)

# 4ï¸â£ Visualize the clusters
plt.figure(figsize=(10, 8))
unique_labels = np.unique(labels)
for label in unique_labels:
    if label == -1:
        # Noise points
        plt.scatter(X_tsne[labels == label, 0], X_tsne[labels == label, 1],
                    label="Noise", c='gray', s=40, alpha=0.6)
    else:
        plt.scatter(X_tsne[labels == label, 0], X_tsne[labels == label, 1],
                    label=f"Cluster {label}", s=40, alpha=0.8)

plt.title("DBSCAN Clustering on t-SNE Reduced Digits Data")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend()
plt.grid(True)
plt.show()
```

34. **Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1ï¸â£ Load and scale the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2ï¸â£ Compute inertia for K = 2 to 6
inertia_values = []
K_range = range(2, 7)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# 3ï¸â£ Plot inertia values
plt.figure(figsize=(8, 6))
plt.plot(K_range, inertia_values, marker='o', linestyle='-', color='blue')
plt.title("K-Means Inertia for Different Cluster Counts")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.xticks(K_range)
plt.show()
```
